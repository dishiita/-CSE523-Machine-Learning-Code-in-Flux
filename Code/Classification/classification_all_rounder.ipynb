```python
import numpy as np                                          #importing the libraries and reading the data
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
from nltk import FreqDist

# original_data=pd.read_csv("data_avg_batsman_runs.csv")
# original_data
original_data= pd.read_csv("ALL_ROUNDERS_AVG_POINTS.csv")


# original_data=original_data[original_data['batsman']=='V Kohli']
# val=original_data['strike_rate'].values
# fdist = FreqDist(val)
# fdist.most_common(2)
# print(val.mean())
# original_data=original_data.head(100)
original_data
```

```python
# plt.style.use('seaborn-whitegrid')
x = original_data["rating"].values    #storing the values of ratings in an array
encoded=[];                         
for w in x :                          #for loop over the rating column
    if(w==9):                         
        encoded.append("9")           #if w = 9; then encode it as 9  
    elif(w==9.5 ):                    
        encoded.append("9.5")         #if w = 9.5; then encode it as 9.5
    elif(w==10):
        encoded.append("10")          #if w = 10; then encode it as 10
    elif(w==8 ):
        encoded.append("8")           #if w = 8; then encode it as 8
    elif(w==8.5 ):
        encoded.append("8.5")         #if w = 8.5; then encode it as 8.5
    
encoded   
    
#     if(w>=0 and w<=20):
#         encoded.append(1)
#     if(w>20 and w<=49):
#         encoded.append(2)
#     if(w>49 and w<=74):
#         encoded.append(3)
# #     if(w>74 and w<=110):
# #         encoded.append(4)
#     if(w>74):
#         encoded.append(4)
        


```

```python
original_data['good_player']=encoded                                      #replacing the values in the column as the encoded values


le = preprocessing.LabelEncoder()                                         #Converting string labels into numbers

venue_encoded=le.fit_transform(original_data['venue'])                    #this will take the values of the venues in total and return it as encoded labels 
print(venue_encoded)

name_encoded=le.fit_transform(original_data['wicket_keeper_batsman'])     #this will take the values of the wicketkeeper batsman i.e. the role of the player is wicketkeeper batsman in total and return it as encoded labels 
print(name_encoded)

opp_team_encoded=le.fit_transform(original_data['opposition_team'])       #this will take the values of the opposition teams in total and return it as encoded labels 
print(opp_team_encoded)

original_data['venue_encoded']=venue_encoded                              #view the columns after encoding the labels
original_data['name_encoded']=name_encoded
original_data['opp_team_encoded']=opp_team_encoded
```

```python
x=original_data[['name_encoded','venue_encoded','opp_team_encoded','batting_avg_inning_points','batting_avg_venue_points','batting_avg_oppo_points','avg_strike_rate','avg_batsman_runs','bowl_fp_avg','bowl_opp_avg','bowl_ven_avg','bowl_inn_avg','bowl_run_avg','eco_avg','maiden_avg','wicket_avg','bowl_over_avg']]
# x=original_data[['name_encoded','venue_encoded','opp_team_encoded']]
# x=original_data[['name_encoded','venue_encoded','opp_team_encoded','avg_oppo_points','avg_venue_points','avg_inning_points','avg_batsman_runs','avg_strike_rate','avg_home_team_points','avg_total_points','avg_bound_runs']]
y=original_data['good_player']                                                        #taking required features or columns for the model training
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)    #dividing the train (75%) and test data (25%) with random state = 42 for data shuffling

model=GaussianNB()                                                                    #Naive Bayes model for classification
model.fit(x_train,y_train)                                                            #pass the training data in the model to predict the test data

y_test
```

```python
y_pred=model.predict(x_test)      #predicting y_test for x_test using the model
y_pred
```

```python
accuracy=accuracy_score(y_test,y_pred)*100    #printing the test accuracy of the data
accuracy
```

```python
from sklearn.metrics import classification_report, confusion_matrix   #confusion matrix for test data vs predicted data
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

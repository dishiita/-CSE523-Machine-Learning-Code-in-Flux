```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
from nltk import FreqDist

original_data= pd.read_csv("Btsmn_wk_New_features_avg.csv")
original_data
```

```python
# plt.style.use('seaborn-whitegrid')
x = original_data["rating"].values
encoded=[];
for w in x :
    if(w==9):
        encoded.append("9")
    elif(w==9.5 ):
        encoded.append("9.5")
    elif(w==10):
        encoded.append("10")
    elif(w==8 ):
        encoded.append("8")
    elif(w==8.5 ):
        encoded.append("8.5")
    
encoded   
```

```python
original_data['good_player']=encoded


le = preprocessing.LabelEncoder()
# Converting string labels into numbers.

venue_encoded=le.fit_transform(original_data['venue'])
print(venue_encoded)



name_encoded=le.fit_transform(original_data['wicket_keeper_batsman'])
print(name_encoded)


opp_team_encoded=le.fit_transform(original_data['opposition_team'])
print(opp_team_encoded)



original_data['venue_encoded']=venue_encoded
original_data['name_encoded']=name_encoded
original_data['opp_team_encoded']=opp_team_encoded

```

```python
x=original_data[['name_encoded','venue_encoded','opp_team_encoded','avg_oppo_points','avg_venue_points','avg_inning_points','avg_batsman_runs','avg_strike_rate','avg_home_team_points','avg_total_points','avg_bound_runs']]
y=original_data['good_player']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)

model=GaussianNB()
model.fit(x_train,y_train)

y_test
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
accuracy=accuracy_score(y_test,y_pred)*100
accuracy
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.ensemble import RandomForestClassifier
x=original_data[['name_encoded','venue_encoded','opp_team_encoded','avg_oppo_points','avg_venue_points','avg_inning_points','avg_batsman_runs','avg_strike_rate','avg_home_team_points','avg_total_points','avg_bound_runs']]
y=original_data['good_player']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)

rf = RandomForestClassifier(n_estimators=340)
rf.fit(x_train,y_train)

yhat = rf.predict(x_test)
y_pred = rf.predict(x_test)

print(classification_report(y_test, y_pred))
```

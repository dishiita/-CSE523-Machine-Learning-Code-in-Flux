```python
import random             #importing the libraries needed
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.cluster import KMeans 
from sklearn.datasets.samples_generator import make_blobs 
%matplotlib inline
```

```python
df = pd.read_csv("ONLY_WICKET_KEEPER_AVG_POINTS.csv")   #reading the dataset
y=list(df['Stumping'])                                  #creating the list of values in Stumping colmn
df=df[['Stumping','strike_rate','catch']]               #viewing selected columns
# y=list(df['fp_points'])
# # df = df[['batsman_runs','strike_rate','economy','wickets']]
# # df.tail(20)
# print(y.count(0))
# print(len(y))
df
```

```python
from sklearn.preprocessing import StandardScaler    #importing StandarScaler
X = df.values[:,0:]                                 #taking all the values from the column
X = np.nan_to_num(X)                                #substitutes any NaN values with numerical values
Clus_dataSet = StandardScaler().fit_transform(X)    #normalizes the data
Clus_dataSet                                        #display the array
```

```python
clusterNum = 5                                                              #cluster number = 5
k_means = KMeans(init = "k-means++", n_clusters = clusterNum, n_init = 15)  #k means clustering for cluster num = 5
k_means.fit(X)                                                              #fitting the clusters
labels = k_means.labels_                                                    #labels for each point in the model
print(labels)                                                               #print the labels
```

```python
points=[]                     #condition for 5 classes
for i in labels:              #for loop for the labels for each point obtained above
    if(i==0):                 #if i=0, then it falls in class 9
        points.append(9)
#   elif(i==9):
#     points.append(8)
    elif(i==1):
        points.append(8)      #if i=1, then it falls in class 8
#   elif(i==8):
#     points.append(10)
#   elif(i==7):
#     points.append(8.5)
    elif(i==2):
        points.append(10)     #if i=2, then it falls in class 10
    elif(i==3):
        points.append(8.5)    #if i=3, then it falls in class 8.5
    elif(i==4):
        points.append(9.5)    #if i=4, then it falls in class 9.5
#   elif(i==4):
#     points.append(9.5)
#   elif(i==6):
#     points.append(9.5)
 
print(points)                 #printing the points
```

```python
df_new = df                 #creating new dataset and storing the points
df_new["Points"] = points
# df_new.to_csv("all_rnd_points.csv")
```

```python
df_new.groupby(points).mean()   #viewing the dataset where it is grouped by points and mean is taken
```

```python
from mpl_toolkits.mplot3d import Axes3D                   #plotting the clusters obtained
fig = plt.figure(1, figsize=(8, 6))
plt.clf()
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)  

plt.cla()
ax.set_xlabel('strike_rate')
ax.set_zlabel('catch')
ax.set_ylabel('Stumping')


y=ax.scatter(X[:,1] ,X[:,0],X[:,2],c= points)             
cbar=plt.colorbar(y,pad=0.10,fraction=0.02,shrink=50.0)
```
